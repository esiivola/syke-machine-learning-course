## Machine learning terminology

- **Features**/**Variables**: Feature/variable is an individual measurable or observable property or characteristic of the phenomenon we are interested in. E.g. in the apartment price example features might be the location or the size of the apartment. The below image illustrates how features are usually presented in machine learning. Different charasteristics can also be modeled as different types. For example, the location of an apartment can for be represented as continuous coordinates (latitude + longitude) or as a zip number. (In classification the variable we are trying to predict is sometimes called "class".)
<img src="img/feature-type.jpg" width="600"/>
- **Dataset**: Data set consists of data points. Each data point consists of atleast one feature. For example in the apartment price example if we can collect the sizes and locations of multiple apartments, we would have a data set. In addition to data, data set *usually* also contains metadata. This metadata usually contains the names of the features and sometimes for example the date when the data set was created.  The below image illustrates a data set constructed from simple drawings. The data set contains of eight data points each with three features (color, shape and size). The image has the original drawings on the left and the data set on the right. Usually data sets are organized as matrices; rows represent data points, columns represent features and cells represent the features of data points
<img src="img/data.jpg" width="400"/>

- **Algorithm**: Algorithms are the work horse of machine learning. Different problems require different algorithms and there is no single (right) algorithm to any problem. However, different models make different assumpions about the problem so it is important to know for what purpose each algorithm has been made for. There is one important point in all algorithms: if the data has low quality, it does not really matter how good of an algorithm you choose. Often this is referred as "garbage in, garbage out".

- **Prediction**: In the ML context the term prediction refers to the estimating the value of the variable of interest, and it does not necessarily refer to prediction of the *future*. An algorithm could for example try to *predict* the species of a measured zooplankton individual, or the volume of tree trunks in a forest. 

## What kind of problems can we solve with (classical) machine learning?

In order to do classical machine learning, we need two things:
- Enough data that is represented as numbers (e.g. how tall is this tree?) or discrete classes (is this tree a spruce or a pine?) or both (this tree is a 20m tall spruce)
- A clearly defined question: What do we want to learn from the data?


The classic machine learning can be divided to two subcategories based on the question we are interested in:

### Supervised machine learning
Supervised learning means learning the relationship of variables that have been measured in the data, for example learning how the house price depends on all the other variables that have been recorded (house location, size, construction year, etc.). The word “supervised” here means that we have examples of the “right answer” to our question, and we want to learn to generalize based on that data, i.e. predict the selling prices of apartments outside of our data.

There are two basic types of the prediction task:

- Predict a numerical value: **Regression**	
    - Predict the age of a tree based on measurements such as the height, circumference, location, species…
    - Predict the chl-a concentration of the lake water given season, temperature, nutrient concentrations…

- Predict a category: **Classification**
    - Predict the plankton taxon based on measurements such as size and shape
    - Predict the presence/absence of an endangered species based on the characteristics of the environment
    - There can be any number of predicted classes, e.g. presence/absence (2 classes) or zoo/phytoplankton taxa of the Baltic Sea (tens of classes)
   
### Unsupervised machine learning
Unsupervised learning means that we do not have the examples of “right answers”, but we want to explore the data and find patterns or associations in it, such as which observations are more similar to each other, or which variable values often go together with each other.
- Divide the observations into groups by similarity: **Clustering**
    - Can we divide lakes into groups based on their water chemistry characteristics such as turbidity, colour, and nutrient concentrations?
- Find relations in data: **Association**
    - E.g. data set might show that meadows that host ladybirds and small tortoiseshell butterflies also often host old world swallowtail butterflies.
- Represent data in a more efficient way: **Dimensionality reduction**   
    - Reduce the number of features (data dimensions) while retaining the essential properties of the data  
    - PCA is a well-known example of dimensionality reduction

## Get the Baltic Sea water quality data from SYKE servers

See details of the API from [here](http://rajapinnat.ymparisto.fi/api/veslabnirajapinta/1.0/)

### First, let us define the parameters to get the data with

```python
# First, define the bounds of the data
params = {'latBegin':'59.00000', 'latEnd':'66.00000', 
          'lonBegin':'17.00000', 'lonEnd':'31.00000',
          'dateBegin':'1950-01-01', 'dateEnd':'2021-12-31'}

bounding_box = [float(params['lonBegin']), float(params['latBegin']), float(params['lonEnd']), float(params['latEnd'])]

#We have "Wrapped" plotting stuff inside file plotting_utils.py, take a look at the code if you are interested
_ = plotting_utils.plot_region_in_map(bounding_box, 
                                      bounding_box_context=plotting_utils.BOUNDS_BALTIC_SEA, 
                                      title="The region where the data is gathered")
```

### Then get the data

```python
url = "http://rajapinnat.ymparisto.fi/api/VeslaBNIRajapinta/1.0/csv?" + urllib.parse.urlencode(params)
s = requests.get(url).content
# The response is in csv format, you can go see the data yourself 
print("The data available here: " + url)

data_frame_raw = pd.read_csv(io.StringIO(s.decode('utf-8')))

# The data description shows that columns TEMP, SALIN, TOTOXY,
# PO4P, TOTP, SIO4, NO3N, NO2N, NO23N, NH4N, TOTN and CHL
# are measurement.
# Rest of the columns,
# LATITUDE, LONGITUDE, OBSDATE, OBSTIME, SHIP, OBSDEP
# are features, something we can decide, measurements are something we
# measure for the selected features
# To make the data more managable, let us assume that we are interested
# in temperature, salinity, oxygen, phosphorus and chlorophyl

# Let us remove the data that does not meet the quality standards:
# This equals to removing the missing features. Another strategy here
# would be to assess the quality and maybe input the features for
# the bas quality data.
# Here, as there is plenty of data, removing is the simplest choice.
# We acknowledge that this might induce bias to the data, but
# consider the risk of this to be small.
quality_cols = ['QTEMP', 'QSALIN', 'QTOTOXY', 'QTOTP', 'QCHL']
for col in quality_cols:
    data_frame_raw = data_frame_raw[data_frame_raw[col] == 1]


# Let us then remove columns we are not interested in
cols_extra = ['ID', 'QTEMP', 'QSALIN', 'QTOTOXY', 'QPO4P', 'QTOTP', 'QSIO4',
              'QNO3N', 'QNO2N', 'QNO23N', 'QNH4N', 'QTOTN', 'QCHL',
              'PO4P', 'SIO4', 'NO3N', 'NO2N', 'NO23N', 'NH4N', 'TOTN']
    
#Now we can remove the quality codes:
data_frame_raw = data_frame_raw.drop(cols_extra, axis=1)

print("5 first rows of the raw data:")
print(data_frame_raw.head())
print("The data has {} features and {} data points".format(data_frame_raw.shape[1], data_frame_raw.shape[0] ))
```

## Acquire the ice thickness data in a similar manner

Use the following code to acquire the data, and save it to 'datasets/ice_thickness_raw.pkl' similarly as we saved water quality data.

```python
# if you are not using the CSC environment and are running the files locally using SYKE VPN, uncomment the follofing:
#os.environ['http_proxy'] = 'http://gate102.vyh.fi:81'
#os.environ['https_proxy'] = 'http://gate102.vyh.fi:81'

# A helper function for the API
def retrieve_all_items(service_url): #https://stackoverflow.com/questions/63775378/python-how-to-extract-data-from-odata-api-that-contains-pages-odata-nextlink
    all_responses = []
    url = service_url
    while True:
        if url is None:
            break
        try:
            response = urllib.request.urlopen(url)
            json_data = json.loads(response.read())
            all_responses = all_responses+ json_data['value']
            url = json_data.get('odata.nextLink', None)
        except exception as e:
            url = None
    return pd.DataFrame(all_responses)

# Retrieve all data realated to ice thickness:
SERVICE_URL = 'http://rajapinnat.ymparisto.fi/api/Hydrologiarajapinta/1.1/odata/'
paikka_df = retrieve_all_items(SERVICE_URL + "Paikka")
lippu_df = retrieve_all_items(SERVICE_URL + "Lippu")
mittaus_df = retrieve_all_items(SERVICE_URL + "Jaanpaksuus")

# Combine data to a single data frame
all_data_df = mittaus_df.merge(paikka_df, how='inner', left_on='Paikka_Id', right_on='Paikka_Id')
all_data_df = all_data_df.merge(lippu_df, how='inner', left_on='Lippu_Id', right_on='Lippu_id')

# Remove snow depth measurements ("lumen syvyys")
all_data_df = all_data_df[all_data_df['Lippu_id']==41]

# Select only the relevant columns from the raw data:
df_small = all_data_df[["Arvo", "Aika", "KoordLat", "KoordLong"]]

#coordinates are in strings that are formatted as ABCDEF, which is equal to AB degrees, CD minutes and EF seconds
#We need to transform them to floats for them to be usable
df_small["KoordLat"] = df_small["KoordLat"].apply(lambda x: float(x[:2]) + float(x[2:4])/60 + float(x[4:])/3600).values
df_small["KoordLong"] = df_small["KoordLong"].apply(lambda x: float(x[:2]) + float(x[2:4])/60 + float(x[4:])/3600).values

print(df_small.head())
```
